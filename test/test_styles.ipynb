```python
import numpy as np
import ziaplot as zp
```

```python
css = '''
* {
    font_size: 14;
}
Point.Text {
    font_size: x-large;
}
'''
with zp.Graph().xrange(0, 10).yrange(0, 12).size(400,400):#.css(css) as g:
    line = zp.Line((0,0), 2)
    zp.Point.at(line, 0).label('14', 'E').style('font_size: 14;')
    zp.Point.at(line, 1).label('xx-large', 'E').style('font_size: xx-large;')
    zp.Point.at(line, 2).label('x-large', 'E').style('font_size: x-large;')
    zp.Point.at(line, 3).label('large', 'E').style('font_size: large;')
    zp.Point.at(line, 4).label('medium', 'E').style('font_size: medium;')
    zp.Point.at(line, 5).label('small', 'E').style('font_size: medium;')
    zp.Point.at(line, 6).label('x-small', 'E').style('font_size: x-small;')
    zp.Point.at(line, 7).label('xx-small', 'E').style('font_size: xx-small;')
```

```python
css = '''
* {
    font_size: 14;
}
Point.Text {
    font_size: xx-large;
}
'''
with zp.Graph().xrange(0, 10).yrange(0, 12).size(400,400).css(css) as g:
    line = zp.Line((0,0), 2)
    zp.Point.at(line, 4).label('Point', 'E')
    zp.Point.at(line, 2).label('normal', 'E').style('font_size: normal;')  # reverts to 14-point
    zp.Point.at(line, 1).label('14', 'E').style('font_size: 14;')
```

```python
# Style tester 
def teststyle():
    with zp.LayoutGrid(columns=2) as fig:
        zp.Pie.fromdict({'a':3, 'b':2, 'c':3, 'd': 2, 'e':4, 'f':2, 'g':4,
                         'h':3, 'i':2, 'j':3}).title('Pies').legend('none')
        with zp.GraphQuad().title('Functions').xrange(-5, 5).yrange(-80, 80):
            zp.Function(lambda x: x**2, xrange=(-5, 5)).endmarkers()
            zp.Function(lambda x: x**3/2, xrange=(-5, 5)).endmarkers()
            zp.Function(lambda x: -x**4/20, xrange=(-5, 5)).endmarkers()
            zp.Point((2, 40)).guidex().guidey().label('A', 'N')
        with zp.Graph().title('Dots').axesnames('x', 'y'):
            x = np.linspace(10, 20, 15)
            y = np.log(x)
            zp.Xy(x, y)
            zp.Xy(x, y*2)
            zp.Xy(x, y*3)
            zp.Xy(x, y*4)
        with zp.BarChartGrouped(groups=('a', 'b', 'c', 'd')).axesnames('Letter', 'Value').title('Bars'):
            zp.BarSeries(2, 2, 4, 3)
            zp.BarSeries(2, 3, 1, 4)
```

```python
teststyle()
```

```python
zp.theme('taffy')
teststyle()
```

```python
zp.theme('pastel')
teststyle()
```

```python
zp.theme('bold')
teststyle()
```

```python
zp.theme('dark')
teststyle()
```

```python
zp.theme('darktaffy')
teststyle()
```

```python
zp.theme('darkbold')
teststyle()
```
