```python
import numpy as np
import ziaplot as zp
```

```python
# Style tester 
def teststyle():
    with zp.GridLayout(columns=2) as fig:
        zp.Pie.fromdict({'a':3, 'b':2, 'c':3, 'd': 2, 'e':4, 'f':2, 'g':4,
                         'h':3, 'i':2, 'j':3},
                         title='Pies', legend='none')
        with zp.XyGraph(title='Functions'):
            zp.Function(lambda x: x**2).endmarkers()
            zp.Function(lambda x: x**3/2).endmarkers()
            zp.Function(lambda x: -x**4/20).endmarkers()
        with zp.XyPlot(title='Dots', xname='x', yname='y'):
            x = np.linspace(10, 20, 15)
            y = np.log(x)
            zp.Xy(x, y)
            zp.Xy(x, y*2)
            zp.Xy(x, y*3)
            zp.Xy(x, y*4)
        with zp.BarChartGrouped(groups=('a', 'b', 'c', 'd'), xname='Letter', yname='Value', title='Bars'):
            zp.BarSeries(2, 2, 4, 3)
            zp.BarSeries(2, 3, 1, 4)
```

```python
teststyle()
```

```python
zp.styles.setdefault(zp.styles.Taffy)
teststyle()
```

```python
zp.styles.setdefault(zp.styles.Pastel)
teststyle()
```

```python
zp.styles.setdefault(zp.styles.Bold)
teststyle()
```

```python
zp.styles.setdefault(zp.styles.Dark)
teststyle()
```

```python
zp.styles.setdefault(zp.styles.DarkTaffy)
teststyle()
```

```python
zp.styles.setdefault(zp.styles.DarkBold)
teststyle()
```
