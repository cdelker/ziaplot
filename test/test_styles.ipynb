```python
import numpy as np
import ziaplot as zp
```

```python
# Style tester 
def teststyle():
    p = zp.Pie(title='Pies').wedges(3,2,3,2,4,2,4,3,2,3).names('a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j')
    p2 = zp.XyGraph(title='Functions')
    p2 += zp.Function(lambda x: x**2).endmarkers()
    p2 += zp.Function(lambda x: x**3/2).endmarkers()
    p2 += zp.Function(lambda x: -x**4/20).endmarkers()

    x = np.linspace(10, 20, 15)
    y = np.log(x)
    p3 = zp.XyPlot(title='Dots', xname='x', yname='y')
    p3 += zp.Xy(x, y)
    p3 += zp.Xy(x, y*2)
    p3 += zp.Xy(x, y*3)
    p3 += zp.Xy(x, y*4)

    p4 = zp.BarChartGrouped(('a', 'b', 'c', 'd'), xname='Letter', yname='Value', title='Bars')
    p4.bar_series((2, 2, 4, 3))
    p4.bar_series((2, 3, 1, 4))

    l1 = zp.Hlayout(p3, p2)
    l2 = zp.Hlayout(p, p4)
    fig = zp.Vlayout(l1, l2)
    return fig
```

```python
teststyle()
```

```python
zp.styles.setdefault(zp.styles.Taffy)
teststyle()
```

```python
zp.styles.setdefault(zp.styles.Pastel)
teststyle()
```

```python
zp.styles.setdefault(zp.styles.Bold)
teststyle()
```

```python
zp.styles.setdefault(zp.styles.Dark)
teststyle()
```

```python
zp.styles.setdefault(zp.styles.DarkTaffy)
teststyle()
```

```python
zp.styles.setdefault(zp.styles.DarkBold)
teststyle()
```
