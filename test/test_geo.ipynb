```python
p1 = (3, 1)
p2 = (2.2, 5)
with zp.Graph().xrange(0, 10).yrange(0, 10).size(400,400) as g:
    zp.Line.from_points(p1, p2)
    zp.Point(p1).label(str(p1), 'NE')
    zp.Point(p2).label(str(p2), 'E')
    line1 = zp.Line(p2, slope=-1)
    line2 = line1.normal(line1.xy(3))
    line3 = line2.normal(line2.xy(6))
    zp.Point.at(line3, 7).label('line3 at x=7')
    zp.Point.at_y(line2, 8).label('line2 at y=8', 'N')
    zp.Point.at_intersection(line2, line3, 4, 8).label('line2 x line3', 'N')
```

```python
def f(x):
    return x**3 / 20 + x**2 / 2

with zp.Graph().axesnames(x='Hey').equal_aspect().xrange(-15, 15).yrange(-5, 15) as g:
    ff = zp.Function(f, xrange=(-10, 5))
    zp.Point.at(ff, -2)
    ff.tangent(x=-8)
    k = 1
    ff.tangent(k).trimd(k, 2, 1)
    ff.normal(k).trimd(k, 2, 1)
```

```python
css = '''
Element {
    colorcycle: black;
}
Angle {
    radius: 30;
    font_size: 14;
}    
'''

with (zp.Diagram().css(css)
          .xrange(0, 1).yrange(0, math.sqrt(3)/2)
          .size(400, 400*math.sqrt(3)/2)) as g:
    a = zp.Segment((0, 0), (1, 0))
    b = zp.Segment((1, 0), (.5, math.sqrt(3)/2))
    c = zp.Segment((.5, math.sqrt(3)/2), (0, 0))
    a.midmarker('|')
    b.midmarker('|')
    c.midmarker('|')
    zp.Angle(a, b, quad=2).label('60')
    zp.Angle(b, c, quad=3).label('60')
    zp.Angle(a, c, quad=4).label('60')
```

```python
with (zp.Diagram().css(css)
          .xrange(0, 1).yrange(0, math.sqrt(3)/2)
          .size(400, 400*math.sqrt(3)/2)) as g:
    a = zp.Segment((0, 0), (1, 0))
    b = zp.Segment((1, 0), (.5, math.sqrt(3)/2))
    c = zp.Segment((.5, math.sqrt(3)/2), (0, 0))
    a.midmarker('|||')
    b.midmarker('||')
    c.midmarker('|')
    zp.Angle(a, b, quad=2)
    zp.Angle(b, c, quad=3, arcs=3)
    zp.Angle(a, c, quad=4, arcs=2)
```

```python
with zp.Diagram().css(css).xrange(0, 1).yrange(0, .6).size(300,200):
    c = zp.Segment((0, 0), (1, 0)).label('c', .5, 'S', color='blue')
    a = zp.Segment((1, 0), (.7, .6)).label('a', .5, 'E', color='blue')
    b = zp.Segment((.7, .6), (0, 0)).label('b', .5, 'NW', color='blue')
    zp.Angle(b, c, quad=4).color('red').label(r'$\alpha$', color='red')
    zp.Angle(a, c, quad=2).color('red').label(r'$\beta$', color='red')
    zp.Angle(a, b, quad=3).color('red').label(r'$\gamma$', color='red')
```
