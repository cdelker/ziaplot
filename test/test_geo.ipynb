```python
import math
import numpy as np
import ziaplot as zp
```

```python
p1 = (3, 1)
p2 = (2.2, 5)
with zp.AxesGraph().xrange(0, 10).yrange(0, 10) as g:
    g.style.canvash = 400
    g.style.canvasw = 400
    zp.Line.from_points(p1, p2)
    zp.Point(*p1).label(str(p1), 'NE')
    zp.Point(*p2).label(str(p2), 'E')
    line1 = zp.Line(p2, slope=-1)
    line2 = zp.Normal.to(line1, 3)
    line3 = zp.Normal.to(line2, 6)
    zp.Point.at(line3, 7).label('line3 at x=7')
    zp.Point.at_y(line2, 8).label('line2 at y=8', 'N')
    zp.Point.at_intersection(line2, line3, 4, 8).label('line2 x line3', 'N')
```

```python
with zp.AxesPlot().xrange(4, 20).yrange(40, 56) as g:
    g.style.canvash = 300
    ff = zp.Line((8,48), .5)
    ff2 = zp.Function(lambda x: .25*(x-12)**2 + 46, xrange=(5, 19)).color('blue')
    p1 = zp.Point.at_intersection(ff, ff2, 6, 10).guidex().guidey()
    p2 = zp.Point.at_intersection(ff, ff2, 16, 18).guidex().guidey()
    zp.IntegralFill.intersection(ff, ff2, x1=6, x2=18).color('blue').alpha(.1)
    
```

```python
def f(x):
    return x**3 / 20 + x**2 / 2

with zp.AxesPlot(xname='Hey').equal_aspect().xrange(-15, 15).yrange(-5, 10) as g:
    g.style.canvash = 400
    g.style.canvasw = 800
    ff = zp.Function(f, xrange=(-10, 5))
    zp.Point.at(ff, -2)
    zp.Tangent.to(ff, x=-8)
    k = 1
    zp.TangentSegment.to(ff, k, 2, 1)
    zp.NormalSegment.to(ff, k, 2, 1)
```

```python
def f(x):
    return x**3 / 20 + x**2 / 2

with zp.AxesPlot(xname='Hey').equal_aspect().xrange(-15, 15).yrange(-5, 10) as g:
    g.style.canvash = 400
    g.style.canvasw = 800
    ff = zp.Function(f, xrange=(-10, 5))
    zp.Point.at_maximum(ff, -10, -5)
    zp.Point.at_minimum(ff, -2, 2)
```

```python
with zp.AxesGraph().xrange(-7, 4).yrange(-3, 5) as g:
    g.style.canvasw = 400
    g.style.canvash = 400
    p1 = (-6,4)
    p3 = (-2,-2)
    p2 = (3,3.5)
    b = zp.Curve(p1, p2, p3, t=.5)
    t = .5
    zp.NormalSegment.to_bezier(b, t)
    zp.TangentSegment.to_bezier(b, t)
    zp.Point.on_bezier(b, .33)

    zp.Point(*p1)
    zp.Point(*p2)
    zp.Point(*p3)
```

```python
with zp.AxesPlot().xrange(-2, 5).yrange(-1, 4).equal_aspect():
    b = zp.BezierCubic((0,0), (0, 4), (6,5), (4, 1)).stroke(':')
    t = .2
    zp.NormalSegment.to_bezier(b, t)
    zp.TangentSegment.to_bezier(b, t)
    zp.Point.on_bezier(b, t)
```

```python
with zp.AxesPlot().xrange(-20, 20).yrange(-20, 20).equal_aspect():
    slope = -3
    line1 = zp.Line((0, 0), slope).color('black')
    line2 = zp.Line((0, 0), -1/slope).color('black')
    zp.Angle(line1, line2, quad=1).label(r'$\pi/2$')
    zp.Angle(line1, line2, quad=2).label('Q2')
    zp.Angle(line1, line2, quad=3).label('Q3')
    zp.Angle(line1, line2, quad=4).label('Q4')
```
