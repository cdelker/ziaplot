Some inspiration for ziaplot came from the TKZ-grapheur package for LaTeX. Here are ziaplot's renderings of some examples in its user manual:
https://ctan.math.utah.edu/ctan/tex-archive/graphics/pgf/contrib/tkz-grapheur/doc/tkz-grapheur-doc-en.pdf

```python
import math
import ziaplot as zp
```

```python
with (zp.AxesGraph()
          .xticks(zp.ticker[0:1:.1])
          .yticks(zp.ticker[0:1:.1])) as g:
    g.style.canvash = 400
    g.style.canvasw = 400
    f1 = zp.Function(lambda x: x*math.exp(x-1), (0,1)).color('red')
    f2 = zp.Line.from_slopeintercept(1, 0).color('teal')
    zp.IntegralFill(f1, f2, 0, 1).color('gray')
    zp.Point.at(f1, .8).guidex().guidey()
    zp.Point.at(f1, .9).guidex().guidey()
    zp.Point.at(f1, .7).guidex().guidey()
```

```python
(zp.AxesGraph()
    .xrange(0, math.pi + math.pi/6)
    .xticks((math.pi/6, math.pi/4, math.pi/3, math.pi/2,
             2*math.pi/3, 3*math.pi/4, 5*math.pi/6, math.pi),
            (r'$\frac{\pi}{6}$', r'$\frac{\pi}{4}$',
             r'$\frac{\pi}{3}$', r'$\frac{\pi}{2}$',
             r'$\frac{2\pi}{3}$', r'$\frac{3\pi}{4}$',
             r'$\frac{5\pi}{6}$', r'$\pi$'),
            minor=zp.ticker[0:28*math.pi/24:math.pi/24])
    .yrange(-1.1, 1.1)
    .yticks((-1, -math.sqrt(3)/2, -math.sqrt(2)/2,
             -.5, 0, .5,
             math.sqrt(2)/2, math.sqrt(3)/2, 1),
            ('$-1$', r'$\sqrt{3}/2$', r'$-\sqrt{2}/2$',
             '$-1/2$', '$0$', '$1/2$',
             r'$\sqrt{2}/2$', r'$\sqrt{3}/2$', '$1$'),
           minor=zp.ticker[-1:1:.05])
            
     )
```

```python
with (zp.AxesGraph()
          .xrange(-7,4)
          .yrange(-3,5)
          .xticks(zp.ticker[-8:5:1])
          .yticks(zp.ticker[-4:6:1])) as g:
    g.style.canvasw = 400
    g.style.canvash = 400
    p1 = (-4, 3)
    p2 = (2, 0)
    p3 = (0, -1)
    zp.Segment(p1, p2).color('blue').trim(-5, 3)
    zp.Point(*p1).color('blue')
    zp.Point(*p2).color('blue')
    zp.Line.from_points((-2, -1), (2, 4)).color('red')
    zp.Line(p3, .25).color('olive')
    zp.Point(*p3).color('olive')
```

```python
with (zp.AxesPlot()
          .xticks(zp.ticker[4:20:1])
          .yticks(zp.ticker[40:56:2])) as g:
    f = zp.Function(lambda x: -2*x+3+24*math.log(2*x), (5, 19))
    for k in [6, 7, 8, 9, 10]:
        zp.Point.at(f, k).guidex().guidey()
```

```python
with (zp.AxesPlot()
          .xticks(zp.ticker[4:20:1])
          .yticks(zp.ticker[40:56:2])) as g:
    f = zp.Function(lambda x: -2*x+3+24*math.log(2*x), (5, 19)).color('red')
    s = zp.Secant.to_function(f, 7.5, 18).color('teal')
    zp.Point.at_intersection(f, s, 7, 8)
    zp.Point.at_intersection(f, s, 17, 19)
```

```python
with (zp.AxesPlot()
          .xticks(zp.ticker[4:20:1])
          .yticks(zp.ticker[40:56:2])) as g:
    f = zp.Function(lambda x: -2*x+3+24*math.log(2*x), (5, 19)).color('red')
    f2 = zp.Function(lambda x: .25*(x-12)**2 + 46, (5, 19)).color('blue')
    zp.Point.at_intersection(f, f2, 6, 8).guidex()
    zp.Point.at_intersection(f, f2, 17, 18).guidex()
```

```python
with (zp.AxesGraph(style=zp.styles.BlackWhite())
          .xrange(-1, 5).yrange(-1, 3)) as g:
    f = zp.Function(lambda x: 0.6*math.cos(4.5*(x-4)+2.1) - 1.2*math.sin(x-4)+.1*x+.2,
                    (.35, 4.2)).color('black')
    zp.Point.at_minimum(f, 1, 2).color('olive').guidex().guidey()
    zp.Point.at_maximum(f, 2, 2.5).color('red').guidex().guidey()
    zp.Point.at_maximum(f, 3, 4).color('blue').guidex().guidey()
```

```python
with zp.AxesGraph().xrange(-7, 4).yrange(-3, 5) as g:
    g.style.canvasw = 400
    g.style.canvash = 400
    p1 = (-6,4)
    p3 = (-2,-2)
    p2 = (3,3.5)
    b = zp.Curve(p1, p2, p3, t=.5)
    zp.Point.on_bezier(b, .33)
    zp.Point(*p1)
    zp.Point(*p2)
    zp.Point(*p3)
```

```python
with zp.AxesPlot().xrange(0, 20).yrange(0, 160):
    ff = zp.Function(lambda x: 80*x*math.exp(-.2*x), xrange=(1,20))
    zp.IntegralFill(ff, x1=2.5, x2=12.5).color('lightblue').alpha(.5)
```

```python
with (zp.AxesPlot()
          .xticks(zp.ticker[4:20:1])
          .yticks(zp.ticker[40:56:2])) as g:
    f = zp.Function(lambda x: -2*x+3+24*math.log(2*x), (5, 19)).color('red')
    s = zp.TangentSegment.to(f, 7.5).color('teal').trim(5, 10)
    zp.Point.at(f, 7.5).color('teal')
```

```python
with (zp.AxesGraph()
          .xrange(0, 10)
          .yrange(0, 16)):
    f = zp.Function(lambda x: -(x-6)**2 + 12).color('teal')
    zp.Secant.to_function(f, 4, 8).color('orange')
    zp.Secant.to_function(f, 3, 6).color('red')
    m = zp.Point.at(f, 6).label('M', 'NW').color('black')
    d = zp.Point.at(f, 4).label('D', 'NW').color('black')
    e = zp.Point.at(f, 8).label('E', 'NE').color('black')
    l = zp.Point.at(f, 3).label('L', 'SE').color('black')
    zp.Segment((8, f.y(8)), (8, 0)).color('blue')
    zp.IntegralFill(f, x1=8, x2=9.5).color('red').alpha(.2)
    zp.TangentSegment.to(f, 5).trim(3, 7).color('plum')
    zp.Point.at(f, 5).color('yellow')
```
