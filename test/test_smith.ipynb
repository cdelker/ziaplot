```python
zp.AxesSmith(grid='extrafine')
```

```python
p = zp.AxesSmith()
p += zp.SmithConstReactance(-.2)
p += zp.SmithConstReactance(1, rmin=.2, rmax=2)
p
```

```python
p = zp.AxesSmith()
p += zp.SmithConstResistance(.2).name('Value 1')
p += zp.SmithConstResistance(1, xmin=-.5, xmax=2).name('Value 2')
p
```

```python
p = zp.AxesSmith()
p += zp.HLine(0)
p += zp.VLine(0)
p += zp.PolyLine([0, .5], [0, .5])
p += zp.LinePolar([0, .2, .4, .6], [135, 145, 155, 165], deg=True)
p
```
