```python
import ziaplot as zp
```

```python
with zp.Diagram().size(400,300).xrange(-.35, 2).yrange(-2, 2) as d:
    P = zp.Bezier((0,0), (1, 2), (2, 0), (3, 0))
    A = zp.Circle((.1,.2), .1).svg.animate_move(P, begin='0', duration='2', rotate='auto')
    B = zp.Segment((0, -.1), (0, .1)).color('blue').svg.animate_move(P, begin='0.5s', duration='2', rotate='auto')
    C = zp.Polygon([(0, -.1), (0, .1), (.1, 0)]).fill('orange').svg.animate_move(P, begin='1s', duration='2', rotate='auto')
    D = zp.Point((0,0)).svg.animate_move(P, begin='1.5s', duration='2', rotate='auto')
```

```python
# Animate In/Out and Set
with zp.LayoutGrid(columns=3):
    with zp.Diagram().size(400,300).xrange(-.1, 2).yrange(-2, 2) as d:
        Q = zp.Circle((0,0), .1)
        P = zp.Segment((0,0), (1, 1))
        P.svg.animate_in(duration='2s', begin='1s')
        P.svg.animate_out(duration='2s', begin='4s', repeat='')
        Q.svg.animate_set('fill', to='red', begin='3s')
        Q.svg.animate_set('fill', to='green', begin='6s')

    with zp.Diagram().size(400,300).xrange(-1, 2).yrange(-2, 2) as d:
        Q = zp.Circle((0,0), .1)
        P = zp.BezierHobby((0,0), (.5, .5), (1, -.5), (2, 0))
        P.svg.animate_in(duration='2s', begin='1s')
        P.svg.animate_out(duration='2s', begin='4s', repeat='')
        Q.svg.animate_set('fill', to='red', begin='3s')
        Q.svg.animate_set('fill', to='green', begin='6s')

    with zp.Diagram().size(400,300).xrange(-1, 2).yrange(-2, 2) as d:
        Q = zp.Circle((1,0), .1)
        P = zp.Ellipse((1, 0), 1, .5)
        P.svg.animate_in(duration='2s', begin='1s')
        P.svg.animate_out(duration='2s', begin='4s', repeat='')
        Q.svg.animate_set('fill', to='red', begin='3s')
        Q.svg.animate_set('fill', to='green', begin='6s')

    with zp.Diagram().size(400,300).xrange(-1, 2).yrange(-2, 2) as d:
        Q = zp.Circle((0,0), .1)
        P = zp.Polygon([(0,0), (1,0), (.5, 1)])
        P.svg.animate_in(duration='2s', begin='1s')
        P.svg.animate_out(duration='2s', begin='4s', repeat='')
        Q.svg.animate_set('fill', to='red', begin='3s')
        Q.svg.animate_set('fill', to='green', begin='6s')

    with zp.Diagram().size(400,300).xrange(-1, 2).yrange(-2, 2) as d:
        Q = zp.Circle((0,0), .1)
        P = zp.Arc((1, 0), .5, 0, 135)
        P.svg.animate_in(duration='2s', begin='1s')
        P.svg.animate_out(duration='2s', begin='4s', repeat='')
        Q.svg.animate_set('fill', to='red', begin='3s')
        Q.svg.animate_set('fill', to='green', begin='6s')

    with zp.Diagram().size(400,300).xrange(-1, 3).yrange(-2, 2) as d:
        Q = zp.Circle((0,0), .1)
        P = zp.Bezier((0,0), (.5, 1), (1, 0))
        P.svg.animate_in(duration='2s', begin='1s')
        P.svg.animate_out(duration='2s', begin='4s', repeat='')
        Q.svg.animate_set('fill', to='red', begin='3s')
        Q.svg.animate_set('fill', to='green', begin='6s')
```
