```python
import ziaplot as zp
```

```python
zp.config.text = 'text'    # Draw text as <text>
p = zp.BarChart(['January', 'February', 'March', 'April'], horiz=True,
                title='Fruits', xname='Month', yname='$N_0$', legend='right')
p.bar((4, 4, 5, 6)).name('Apple')
p.bar((3, 4, 4, 5)).name(r'$\frac{1}{x+1}$')
p.bar((2, 1, 5, 4)).name(r'$\sqrt{y}$')
p.bar((3, 3, 2, 3)).name('Orange')
p
```

```python
zp.config.text = 'path'  # Draw text as <path>
p = zp.BarChart(['$a^2$', '$b^2$', '$c^2$', '$d^2$'], horiz=True,
                title=r'Title, $\frac{F_0}{a+b}$',
                xname=r'X Label $\frac{a + b}{c}$',
                yname=r'$\frac{N_0}{a+b}$',
                legend='right')
p.bar((4, 4, 5, 6)).name('Apple')
p.bar((3, 4, 4, 5)).name(r'$\frac{1}{x+1}$')
p.bar((2, 1, 5, 4)).name(r'$\sqrt{y}$')
p.bar((3, 3, 2, 3)).name('Orange')
#assert 'symbol' in p.svg()
p
```

```python
zp.config.svg2 = False   # Draw text as <path> using SVG1.x
p = zp.BarChart(['$a^2$', '$b^2$', '$c^2$', '$d^2$'], horiz=True,
                title=r'Title, $\frac{F_0}{a+b}$',
                xname=r'X Label $\frac{a + b}{c}$',
                yname=r'$\frac{N_0}{a+b}$',
                legend='right')
p.bar((4, 4, 5, 6)).name('Apple')
p.bar((3, 4, 4, 5)).name(r'$\frac{1}{x+1}$')
p.bar((2, 1, 5, 4)).name(r'$\sqrt{y}$')
p.bar((3, 3, 2, 3)).name('Orange')
assert 'symbol' not in p.svg()
p
```

```python
zp.config.text = 'path'  # Draw text as <path>
zp.config.svg2 = True
x = zp.linspace(0, 10, 20)
y = x
p = zp.XyPlot()
p += zp.Xy(x, y).marker('round', radius=3)
p += zp.Text(2, 4, '(2, 3)')
p += zp.Text(4, 6, '(4, 5)').color('red')
p += zp.Text(4, 2, 'center', halign='center', valign='center')
p += zp.Text(6, 2, 'topleft', halign='left', valign='top')
p += zp.Text(8, 2, 'botright', halign='right', valign='bottom')
p += zp.Text(6, 8, '$ax^2 + bx + c$ rotate', rotate=40).color('blue')
p
```
