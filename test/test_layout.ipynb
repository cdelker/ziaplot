```python
import numpy as np
import ziaplot as zp
```

```python
x = np.linspace(-1, 1)
y = x
y2 = x**2
y3 = x**3
y4 = np.exp(x)
```

```python
# Add Lines/Series directly to Layout
p1 = zp.Line(x, y)
p2 = zp.Line(x, y2)
p3 = zp.Line(x, y3)
p4 = zp.Line(x, y4)

zp.Hlayout(p1, p2, p3, height=200, width=800, sep=-10)
```

```python
zp.Vlayout(p1, p2, p3, height=800, width=300)
```

```python
# Add Lines to Axes, then to Layout (should look the same)
p1 = zp.XyPlot()
p1 += zp.Line(x, y)
p2 = zp.XyPlot()
p2 += zp.Line(x, y2)
p3 = zp.XyPlot()
p3 += zp.Line(x, y3)
p4 = zp.XyPlot()
p4 += zp.Line(x, y4)
zp.Hlayout(p1, p2, p3, p4, height=200, width=1400)
```

```python
# GridLayout
p1 = zp.XyPlot()
p1 += zp.Line(x, y)
p2 = zp.XyPlot()
p2 += zp.Line(x, y2)
p3 = zp.XyPlot()
p3 += zp.Line(x, y3)
p4 = zp.XyPlot()
p4 += zp.Line(x, y4)
zp.GridLayout(p1, p2, p3, p4.span(3), columns=3, width=1000)
```

```python
# Add layouts to layouts
p1 = zp.XyPlot()
p1 += zp.Line(x, y)
p2 = zp.XyPlot()
p2 += zp.Line(x, y2)
p3 = zp.XyPlot()
p3 += zp.Line(x, y3)
p4 = zp.XyPlot()
p4 += zp.Line(x, y4)
zp.GridLayout(p1.span(1,2), p2, p3, p4.span(2), columns=2)
```

```python
# Layouts as context managers
x = np.linspace(0, 2*np.pi)
y = np.sin(x)

with zp.GridLayout(columns=2):
    with zp.XyPlot():
        zp.Line(x, y)
        zp.Line(x, y*2)
    zp.Line(x, y**2)
    zp.Pie.fromdict({'a':20,'b':30,'c':40})
    zp.BarChart.fromdict({'A':3, 'B':2, 'C':6})
```
