```python
import numpy as np
import ziaplot as zp
```

```python
x = np.linspace(-1, 1)
y = x
y2 = x**2
y3 = x**3
y4 = np.exp(x)
```

```python
# Add Lines/Series directly to Layout
p1 = zp.Line(x, y)
p2 = zp.Line(x, y2)
p3 = zp.Line(x, y3)
p4 = zp.Line(x, y4)

zp.Hlayout(p1, p2, p3, height=200, width=800, sep=-10)
```

```python
zp.Vlayout(p1, p2, p3, height=800, width=300)
```

```python
# Add Lines to Axes, then to Layout (should look the same)
p1 = zp.XyPlot()
p1 += zp.Line(x, y)
p2 = zp.XyPlot()
p2 += zp.Line(x, y2)
p3 = zp.XyPlot()
p3 += zp.Line(x, y3)
p4 = zp.XyPlot()
p4 += zp.Line(x, y4)
zp.Hlayout(p1, p2, p3, p4, height=200, width=800, sep=-10)
```

```python
# Add layouts to layouts - can add the same axis more than once
row1 = zp.Hlayout(p1, p2, p3)
row2 = zp.Hlayout(p4, p4)
zp.Vlayout(row1, row2)
```

```python
# Add layouts to layouts - can add the same axis more than once
col1 = zp.Vlayout(p1, p2, p3)
col2 = zp.Vlayout(p4, p4)
zp.Hlayout(col1, col2)
```
