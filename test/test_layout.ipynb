```python
import numpy as np
import ziaplot as zp
```

```python
x = np.linspace(-1, 1)
y = x
y2 = x**2
y3 = x**3
y4 = np.exp(x)
```

```python
# Add Lines/Series directly to Layout
p1 = zp.Plot(x, y)
p2 = zp.Plot(x, y2)
p3 = zp.Plot(x, y3)
p4 = zp.Plot(x, y4)

zp.LayoutH(p1, p2, p3, height=200, width=800, sep=-10)
```

```python
zp.LayoutV(p1, p2, p3, height=800, width=600)
```

```python
# Add Lines to Axes, then to Layout (should look the same)
p1 = zp.Graph()
p1 += zp.Plot(x, y)
p2 = zp.Graph()
p2 += zp.Plot(x, y2)
p3 = zp.Graph()
p3 += zp.Plot(x, y3)
p4 = zp.Graph()
p4 += zp.Plot(x, y4)
zp.LayoutH(p1, p2, p3, p4, height=200, width=1400)
```

```python
# LayoutGrid
p1 = zp.Graph()
p1 += zp.Plot(x, y)
p2 = zp.Graph()
p2 += zp.Plot(x, y2)
p3 = zp.Graph()
p3 += zp.Plot(x, y3)
p4 = zp.Graph()
p4 += zp.Plot(x, y4)
zp.LayoutGrid(p1, p2, p3, p4.span(3), columns=3, width=1000)
```

```python
p1 = zp.Graph()
p1 += zp.Plot(x, y)
p2 = zp.Graph()
p2 += zp.Plot(x, y2)
p3 = zp.Graph()
p3 += zp.Plot(x, y3)
p4 = zp.Graph()
p4 += zp.Plot(x, y4)
zp.LayoutGrid(p1.span(1,2), p2, p3, p4.span(2), columns=2)
```

```python
# column_widths and row_heights
with zp.LayoutGrid(columns=3, column_gap=0, column_widths='2fr 1fr 1fr',
                   row_gap=10, row_heights='1fr 2fr') as d:
    zp.Plot([1,2,3], [1,2,5]).span(2)
    with zp.Graph().noyticks() as p1:
        zp.Plot([1,2,3], [3,3,2]).color('blue')
    zp.Plot([1,2,3], [0,2,6]).color('orange')
    zp.Plot([1,2,3], [0,2,6]).color('maroon')
    with zp.Graph().noyticks() as p1:
        zp.Xy([100, 200, 300], [100, 300, 350]).color('green')
    
```

```python
# Layouts as context managers
x = np.linspace(0, 2*np.pi)
y = np.sin(x)

with zp.LayoutGrid(columns=2):
    with zp.Graph():
        zp.Plot(x, y)
        zp.Plot(x, y*2)
    zp.Plot(x, y**2)
    zp.Pie.fromdict({'a':20,'b':30,'c':40})
    zp.BarChart.fromdict({'A':3, 'B':2, 'C':6})
```
